# The Beta and Binomial Distributions

### In a small survey, a random sample of **50** people from a large population is selected. Each person is asked a question to which the answer is either `Yes` or `No`. Let the proportion in the population who would answer `Yes` be **θ**. Our prior distribution for **θ** is a `beta(1.5, 1.5)` distribution. In the survey, **37** people answer `Yes`.

```python
# encoode our prior parameters
α = 1.5
β = 1.5

# visualize what this distribution looks like
x = np.linspace(beta.ppf(0.01, α, β), beta.ppf(0.99, α, β))
prior_plot = plt.plot(x, beta.pdf(x, α, β))
title = plt.title(f"Prior of Beta, α={α}, β={β}")
```

### Find the prior mean and prior standard deviation of **θ**.

\begin{align}
E(\theta) = \frac{\alpha}{\alpha + \beta}
\end{align}

```python
def find_beta_expected_value(α, β):
    return α / (α + β)

prior_expected_value = find_beta_expected_value(1.5, 1.5)
print(f"The prior mean with parameters α={α} and β={β} is {prior_expected_value}")
```

\begin{align}
var(\theta) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}
\end{align}

```python
def find_beta_variance(α, β):
    return (α * β) / (((α + β) **2) * (α + β + 1))

prior_variance = find_beta_variance(1.5,1.5)
print(f"The prior variance with parameters α={α} and β={β} is {prior_variance}")
```

### Find the prior probability that **θ** < 0.6.

```python
cutoff = beta.cdf(.6, α, β)
print(f"Probability of θ being less than 0.6 is {cutoff}.")
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('θ')
ax1.set_ylabel('PDF', color=color)
ax1.plot(x, beta.pdf(x, α, β), color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

SAMPLES = 10000
samples = beta.rvs(α, β, size=SAMPLES)
plt.hist(samples, alpha=0.4, bins=500)
plt.axvline(cutoff)
color = 'tab:blue'
ax2.set_ylabel('Empirical Sample Distribution', color=color)  # we already handled the x-label with ax1
ax2.hist(samples, color=color, alpha=0.5)
ax2.tick_params(axis='y', labelcolor=color)
```

### Find the prior likelihood.

We know that 50 people were sampled, with 37 people answering **Yes**. Therefore, `50 - 37 = 13` people answered **No**. Given our prior distribution on **θ**, the likelihood is a binomial distribution:

\begin{equation*}
P(θ) = {n \choose θ} p^θ (1-p)^{n-θ}
\end{equation*}

Note that here, θ references the **number of successes amongst n trials**, not a probability between 0 and 1. Therefore, in this case, θ = 37, and p = 0.5:

\begin{equation*}
P(37) = {50 \choose 37} 0.5^{37} (1-0.5)^{13}
\end{equation*}

The formula for combinations is defined as follows:

!["here"](combination.png)

Let's define our own function to compute the combinations. We'll first need to define a function to compute the factorial:

```python
def factorial(number):
    if number > 0:
        return number * factorial(number - 1)
    else:
        return 1
```

```python
def combination(total_number, successes):
    return factorial(total_number) / (factorial(successes) * factorial(total_number - successes))
```

Now we can use `combination()` to find the likelihood of our binomial distribution given the prior parameters:

```python
def binomial_likelihood_function(totalnumber, successes, probability):
    return combination(50, 37) * (probability ** 37) * ((1 - probability) ** 13)
    
```

### Find the posterior distribution of θ.

The posterior distribution of θ is defined by

\begin{equation}
E(θ) = \frac{α + successes}{α + β + failures}
\end{equation}

We'll encode this functionality into a Python class to make it more reusable for the future:

```python
class BetaDistribution:
    
    def __init__(self, α=1, β=1):
        self.α = α
        self.β = β
    
    def update(self, successes, total_samples):
        self.α += successes
        self.β += (total_samples - successes)
```

```python
posterior_beta = BetaDistribution(α=1.5, β=1.5) #these were our priors
posterior_beta.update(successes=37, total_samples=50) # update our distribution based on the new evidence (successes, total samples)
```

### Find the posterior mean and posterior standard deviation of θ

Our expected value for a beta distribution:

\begin{align}
E(\theta) = \frac{\alpha}{\alpha + \beta}
\end{align}

Our variance for a beta distribution:

\begin{align}
var(\theta) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}
\end{align}

```python
find_beta_expected_value(posterior_beta.α, posterior_beta.β)
```

```python
find_beta_variance(posterior_beta.α, posterior_beta.β)
```

Notice that the expected value from **0.5** to **0.73** increased, while the variance decreased from **0.06** to **0.0037** (we become more certain in our findings).

### Plot a graph showing the prior and posterior probability density functions of θ on the same axes.

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('θ')
ax1.set_ylabel('Prior Distribution', color=color)
ax1.plot(x, beta.pdf(x, α, β), color=color) # these were our priors
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Posterior Distribution', color=color)  # we already handled the x-label with ax1
ax2.plot(x, beta.pdf(x, posterior_beta.α, posterior_beta.β), color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

### Find the posterior probability that θ < 0.6

```python
beta.cdf(0.6, posterior_beta.α, posterior_beta.β)
```

There is now less than a **2.5%** chance that the true rate of "Yes" is below 60%!

### Given the information that we now know, what is the most likely θ?

This question essentially asks us to find the MAP (***maximum a posteriori***) point estimate of the posterior distribution. This is essentially the point for θ where the pdf of the posterior Beta distribution is the highest (the mode). The mode is given by

\begin{equation}
mode(θ) = \frac{α - 1}{α + β -2}
\end{equation}

```python
def find_mode_of_beta(α=1, β=1):
    return (α - 1) / (α + β - 2)

find_mode_of_beta(posterior_beta.α, posterior_beta.β)
```

# Τhe Gamma and Poisson Distribution

#### The populations, `n`, and the number of cases, `x`, of a disease in a year in each of six districts are given in the table below.

| Population (***n***)  | Cases (***x***) |
| ------------- | ------------- |
| 120,342  | 2  |
| 235,967  | 5  |
| 243,745  | 3 |
| 197,452  | 5 |
| 276,935 | 3  |
| 157,222 | 1 |

#### We suppose that the number `X_i` in a district with population ni is a Poisson random variable with mean ni λ/100000. The number in each district is independent of the numbers in other districts, given the value of λ. Our prior distribution for λ is a gamma distribution with mean 3.0 and standard deviation 2.0.



```python
TOTAL = 120342 + 235967 + 243745 + 197452 + 276935 + 157222
CASES = 2 + 5 + 3 + 5 + 3 + 1

print(f"There are a total of {TOTAL} people in these districts, with {CASES} cases.")
```

### Find the parameters of the prior distribution

The expected value of a gamma distribution is given by

\begin{equation}
E(θ) = \frac{\alpha}{\beta}
\end{equation}

The variance of a gamma distribution is given by

\begin{equation}
var(θ) = \frac{\alpha}{\beta^2}
\end{equation}

We can solve for α and β with some algebra:

\begin{equation}
\frac{E(θ)}{var(θ)} = \frac{3}{2^2} = 0.75 = \frac{\frac{α}{β}}{\frac{α}{β^2}} = \frac{α}{β}\frac{β^2}{α} = β
\end{equation}

Once we find β, we can substitute in its value to find the value of α:

\begin{equation}
E(θ) = 3 = \frac{α}{β} = \frac{α}{0.75}
\end{equation}

\begin{equation}
α = (0.75)(3) = 2.25
\end{equation}

### Find the prior probability that λ < 2.0

Since λ has a prior distribution that is described by `gamma(α=2.25, β=0.75)`, we can use its CDF to find the distribution less than 2.0 (note that if you are using `scipy` gamma distributions, `scale` is the inverse of the β parameter):

```python
gamma.cdf(2, a=2.25, scale=1/0.75)
```

```python
# visualize what this distribution looks like
x = np.linspace(gamma.ppf(0.01, 2.25, 0.75), gamma.ppf(0.99, 2.25, 0.75))
prior_plot = plt.plot(x, gamma.cdf(x, a=2.25, scale=1/0.75))
plt.axvline(2, color="black")
title = plt.title(f"Prior of Gamma, α={2.25}, β={0.75}")
```

### Find the Likelihood Given the Prior Distribution

!["here"](poisson.png)

Since we have many observations, we must factorize these likelihoods:

\begin{equation}
P(x) = \prod\frac{e^{-λ_i} λ^{x_i}}{x_i!}
\end{equation}

Here, $λ_i$ is the λ value from each community, $x_i$ represents the number of disease cases per community. This likelihood can be simplified to

\begin{equation}
P(x) = \frac{e^{\sum -λ_i} λ^{\sum x_i}}{x_i!}
\end{equation}

We can disregard the denominator for right now to get a proportional lieklihood:

\begin{equation}
P(x) \propto e^{\sum -λ_i} λ^{\sum x_i}
\end{equation}

Since we know that x_i will sum to `CASES`, we can substitute in `19` for $\sum x_i$. Furthermore, we can subsitute in 12.13 for $\sum -λ_i$, since this is $\frac{\sum{x_i}}{\sum{n_i}}$:

\begin{equation}
P(x) \propto e^{-12.13} λ^{19}
\end{equation}

### Find the posterior distribution of λ

The prior distribution of λ is given by

![here](gamma.svg)

The first time evaluates to a constant scaling parameter here. Therefore, our original prior distribution can be expressed as 

\begin{equation}
p(θ) \propto x^{α-1}e^{-βx}
\end{equation}

Plugging in our prior $α=2.25$ and $β=.075$, we get 

\begin{equation}
p(θ) \propto x^{2.25-1}e^{-0.75x}
\end{equation}
\begin{equation}
p(θ) \propto x^{1.25}e^{-0.75x}
\end{equation}

Using Bayes' Rule:
\begin{equation}
posterior \propto likelihood * prior
\end{equation}

\begin{equation}
p(θ|X) \propto p(X|θ)p(θ)
\end{equation}

\begin{equation}
p(θ|X) \propto (e^{-0.75λ} λ^{1.25})(e^{-12.31λ} λ^{19})
\end{equation}

We can construct a Python function to update our distribution accordingly:

```python
class PoissonDistribution:
    
    def __init__(self, α, β):
        self.α = α
        self.β = β
    
    def update(self, evidence_rate, count):
        self.α += count
        self.β += evidence_rate
        
poisson = PoissonDistribution(α=2.25, β=0.75)
poisson.update(evidence_rate=12.13, count=19)
```

This is essentially calculating the following:

\begin{equation}
p(θ|X) \propto e^{-13.06λ} λ^{21.25}
\end{equation}

Thus, our new posteriors for **α=21.25** and **β=13.06**.

### Find the posterior mean and posterior standard deviation of λ

The expected value of the gamma distribution is $\frac{\alpha}{\beta}$, and its standard deviation is $\frac{\sqrt{α}}{β}$:

```python
def find_gamma_expected_value(α, β):
    return α / β

def find_gamma_standard_deviation(α, β):
    return np.sqrt(α) / β

print(f"Posterior distribution on λ has mean of {round(find_gamma_expected_value(21.25, 13.06),2)} and \
standard deviation of {round(find_gamma_standard_deviation(21.25, 13.06),2)}")
```

### Plot a graph showing the prior and posterior probability density functions of λ on the same axes.

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('θ')
ax1.plot(x, gamma.pdf(x, 2.25, scale=1/0.75), color=color) # these were our priors
ax1.tick_params(axis='y', labelcolor=color)
color = 'tab:blue'
ax1.set_ylabel('Prior/Posterior Distribution', color=color)  # we already handled the x-label with ax1
ax1.plot(x, gamma.pdf(x, 21.25, scale=1/13.06), color=color)
ax1.tick_params(axis='y', labelcolor=color)
```

# Transitions

Geologists note the type of rock at fixed vertical intervals of six inches up a quarry face. At this quarry there are four types of rock. The following model is adopted. The conditional probability that the next rock type is $j$ given that the present type is $i$ and given whatever has gone before is $p_{ij}$ . Clearly $\sum_{j=1}^{4}p_{ij} = 1$ for all $i$. The following table gives the observed (upwards- **note: anytime you see observed, this is our evidence!**) transition frequencies:

![here](transitions.png)

Our prior distribution for the transition probabilities is as follows. For each $i$ we have a uniform distribution over the space of possible values of $p_{i1},\dots, p_{i4}$. The prior distribution
of $p_{i1},\dots, p_{i4}$ is independent of that for $p_{k1},\dots, p_{k4}$ for $i \neq k$.
