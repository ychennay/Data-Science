# The Beta and Binomial Distributions

### In a small survey, a random sample of **50** people from a large population is selected. Each person is asked a question to which the answer is either `Yes` or `No`. Let the proportion in the population who would answer `Yes` be **θ**. Our prior distribution for **θ** is a `beta(1.5, 1.5)` distribution. In the survey, **37** people answer `Yes`.

```python
# encoode our prior parameters
α = 1.5
β = 1.5

# visualize what this distribution looks like
x = np.linspace(beta.ppf(0.01, α, β), beta.ppf(0.99, α, β))
prior_plot = plt.plot(x, beta.pdf(x, α, β))
title = plt.title(f"Prior of Beta, α={α}, β={β}")
```

### Find the prior mean and prior standard deviation of **θ**.

\begin{align}
E(\theta) = \frac{\alpha}{\alpha + \beta}
\end{align}

```python
def find_beta_expected_value(α, β):
    return α / (α + β)

prior_expected_value = find_beta_expected_value(1.5, 1.5)
print(f"The prior mean with parameters α={α} and β={β} is {prior_expected_value}")
```

\begin{align}
var(\theta) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}
\end{align}

```python
def find_beta_variance(α, β):
    return (α * β) / (((α + β) **2) * (α + β + 1))

prior_variance = find_beta_variance(1.5,1.5)
print(f"The prior variance with parameters α={α} and β={β} is {prior_variance}")
```

### Find the prior probability that **θ** < 0.6.

### Find the prior likelihood.

We know that 50 people were sampled, with 37 people answering **Yes**. Therefore, `50 - 37 = 13` people answered **No**. Given our prior distribution on **θ**, the likelihood is a binomial distribution:

\begin{equation*}
P(θ) = {n \choose θ} p^θ (1-p)^{n-θ}
\end{equation*}

Note that here, θ references the **number of successes amongst n trials**, not a probability between 0 and 1. Therefore, in this case, θ = 37, and p = 0.5:

\begin{equation*}
P(37) = {50 \choose 37} 0.5^{37} (1-0.5)^{13}
\end{equation*}

The formula for combinations is defined as follows:

!["here"](combination.png)

Let's define our own function to compute the combinations. We'll first need to define a function to compute the factorial:

```python
def factorial(number):
    if number > 0:
        return number * factorial(number - 1)
    else:
        return 1
```

```python
def combination(total_number, successes):
    return factorial(total_number) / (factorial(successes) * factorial(total_number - successes))
```

Now we can use `combination()` to find the likelihood of our binomial distribution given the prior parameters:

```python
def binomial_likelihood_function(totalnumber, successes, probability):
    return combination(50, 37) * (probability ** 37) * ((1 - probability) ** 13)
    
```

### Find the posterior distribution of θ.

The posterior distribution of θ is defined by

\begin{equation}
E(θ) = \frac{α + successes}{α + β + failures}
\end{equation}

We'll encode this functionality into a Python class to make it more reusable for the future:

```python
class BetaDistribution:
    
    def __init__(self, α=1, β=1):
        self.α = α
        self.β = β
    
    def update(self, successes, total_samples):
        self.α += successes
        self.β += (total_samples - successes)
```

```python
posterior_beta = BetaDistribution(α=1.5, β=1.5) #these were our priors
posterior_beta.update(successes=37, total_samples=50) # update our distribution based on the new evidence (successes, total samples)
```

### Find the posterior mean and posterior standard deviation of θ

Our expected value for a beta distribution:

\begin{align}
E(\theta) = \frac{\alpha}{\alpha + \beta}
\end{align}

Our variance for a beta distribution:

\begin{align}
var(\theta) = \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}
\end{align}

```python
find_beta_expected_value(posterior_beta.α, posterior_beta.β)
```

```python
find_beta_variance(posterior_beta.α, posterior_beta.β)
```

Notice that the expected value from **0.5** to **0.73** increased, while the variance decreased from **0.06** to **0.0037** (we become more certain in our findings).

### Plot a graph showing the prior and posterior probability density functions of θ on the same axes.

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('θ')
ax1.set_ylabel('Prior Distribution', color=color)
ax1.plot(x, beta.pdf(x, α, β), color=color) # these were our priors
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Posterior Distribution', color=color)  # we already handled the x-label with ax1
ax2.plot(x, beta.pdf(x, posterior_beta.α, posterior_beta.β), color=color)
ax2.tick_params(axis='y', labelcolor=color)
```

### Find the posterior probability that θ < 0.6

```python
beta.cdf(0.6, posterior_beta.α, posterior_beta.β)
```

There is now less than a **2.5%** chance that the true rate of "Yes" is below 60%!

### Given the information that we now know, what is the most likely θ?

This question essentially asks us to find the MAP (***maximum a posteriori***) point estimate of the posterior distribution. This is essentially the point for θ where the pdf of the posterior Beta distribution is the highest (the mode). The mode is given by

\begin{equation}
mode(θ) = \frac{α - 1}{α + β -2}
\end{equation}

```python
def find_mode_of_beta(α=1, β=1):
    return (α - 1) / (α + β - 2)

find_mode_of_beta(posterior_beta.α, posterior_beta.β)
```
